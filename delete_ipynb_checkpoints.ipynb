```python
#Clean up script in Python : delete .ipynb_checkpoints folders
```

```python
import os
import shutil

def remove_checkpoint_folders(root_dir="."):
    """Remove all .ipynb_checkpoints folders"""
    for root, dirs, files in os.walk(root_dir):
        if '.ipynb_checkpoints' in dirs:
            checkpoint_path = os.path.join(root, '.ipynb_checkpoints')
            print(f"Removing: {checkpoint_path}")
            shutil.rmtree(checkpoint_path)
    print("âœ“ All checkpoint folders removed")

# Run it
remove_checkpoint_folders("Data")
```
